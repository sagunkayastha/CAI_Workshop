```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(-10,10,21)
y = x**2
```

```python
x
```

```python
y
```

### Functional matplotlib

```python
plt.plot(x,y)
```

### Labels and Title


```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title(' Title')
```

## Functional Method

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Object oriented method

```python
fig = plt.figure()  # Imaginary blank canvas
#axes = fig.add_axes([left, button, height, width])
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

# Subplots

```python
fig,axes = plt.subplots(nrows =5 , ncols = 2)  #does both plt.figure and axes
plt.tight_layout()
```

```python
axes.shape
```

```python
fig,axes = plt.subplots(nrows =1 , ncols = 2)  #does both plt.figure and axes
for t in axes:
    print(t)
    t.plot(x,y)
plt.show()
```

```python
fig,axes = plt.subplots(nrows =1 , ncols = 2)  #does both plt.figure and axes
axes[0].plot(x,y,'r')
axes[0].set_title('First Plot')
axes[1].plot(y,x,'b')
axes[1].set_title('2nd Plot')
```

```python
fig, axes = plt.subplots(nrows=2,ncols =1, figsize= (8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

## Legends, Labels and titles

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2)
ax.plot(x, x**3)
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='X squared')
ax.plot(x, x**3, label = 'X cubed')
ax.legend(loc=1)
```

# Colors, linewidth, linetypes

### Color

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'black') #basic color names
ax.plot(x,y, color = '#FF8C00')  #RGB hexcode (tools online to get the hex code)
```

### Linewidth and alpha

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 5, alpha=.1)  # try different numbers  #linewidth : lw
```

### Linestyle

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 5, alpha=0.5, linestyle = ':') #linestyle : ls
```

### Markers

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', lw = 1, alpha=0.5, ls = '-', marker = 'o')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'black', lw = 5, alpha=0.5, ls = '-', marker = 'o', markersize=10,
       markerfacecolor='yellow',markeredgewidth=5, markeredgecolor='red')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
